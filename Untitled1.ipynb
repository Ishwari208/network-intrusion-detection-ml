# Network Intrusion Detection using Machine Learning

## Problem Statement
The objective of this project is to develop a supervised machine learning model that classifies network traffic as either normal or malicious using the NSL-KDD dataset.

## Motivation
Traditional rule-based intrusion detection systems struggle to detect novel or evolving attacks. Machine learning enables automated pattern recognition and improved detection performance.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Dataset Description

The NSL-KDD dataset is a benchmark dataset used for evaluating intrusion detection systems.
It contains network connection records with 41 features describing various characteristics of network traffic.

Each record includes:
- Duration of connection
- Protocol type (TCP/UDP/ICMP)
- Service type (HTTP, FTP, etc.)
- Number of bytes transferred
- Error rates and other traffic statistics

Originally, the dataset contains multiple attack categories such as:
- Neptune
- Smurf
- Satan
- Ipsweep

For this project, the problem was simplified into binary classification:
- 0 → Normal Traffic
- 1 → Malicious Attack



```python
import pandas as pd

data = pd.read_csv("KDDTrain+.txt", header=None)

print("Dataset Shape:", data.shape)
data.head()
```

```python
data.iloc[:, -2].value_counts().head(10)
```

## Data Preprocessing

Raw network data cannot be directly used for machine learning.
Therefore, the following preprocessing steps were performed:

1. Converted multi-class attack labels into binary classification:
   - 0 → Normal traffic
   - 1 → Malicious attack

2. Removed unnecessary columns such as the original attack label column
   and difficulty level column.

3. Separated input features (X) and target variable (y).

4. Converted categorical variables (protocol, service, flag)
   into numerical format using One-Hot Encoding.

These preprocessing steps ensure that the dataset is clean, structured,
and suitable for supervised machine learning models.


```python
# Create binary label
data['label'] = data.iloc[:, -2].apply(lambda x: 0 if x == 'normal' else 1)

data['label'].value_counts()
```

```python
# Features (remove original label column and difficulty column)
X = data.iloc[:, :-2]

# Target
y = data['label']

print("Feature shape:", X.shape)
print("Target shape:", y.shape)
```

```python
# Drop original attack label column (index 41)
# Drop difficulty column (index 42)
# Drop our new binary label column from features

X = data.drop(columns=[41, 42, 'label'])

y = data['label']

print("Feature shape:", X.shape)
print("Target shape:", y.shape)
```

```python
X.dtypes.value_counts()
```

```python
X.select_dtypes(include='object').columns
```

```python
# One-hot encode categorical columns
X_encoded = pd.get_dummies(X, columns=[1, 2, 3])

print("Encoded feature shape:", X_encoded.shape)
```

## Train-Test Split

To evaluate model performance properly, the dataset was divided into training and testing sets.

- 80% of the data was used for training the model.
- 20% of the data was used for testing the model.

This ensures that the model is evaluated on unseen data and prevents overfitting.

The random_state parameter was set to ensure reproducibility of results.


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

print("Training set:", X_train.shape)
print("Testing set:", X_test.shape)
```

```python
# Convert all column names to string
X_encoded.columns = X_encoded.columns.astype(str)

# Re-split again (important after modification)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)
```

## Logistic Regression Model

Logistic Regression was implemented as a baseline classification model.

It is a supervised learning algorithm used for binary classification tasks.
It estimates the probability that a given network connection belongs to the malicious class using a linear decision boundary.

This model serves as a simple and interpretable starting point for evaluating intrusion detection performance.


```python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression(max_iter=1000)
model_lr.fit(X_train, y_train)

y_pred_lr = model_lr.predict(X_test)

print("Model trained successfully.")
```

### Logistic Regression Evaluation

The performance of the Logistic Regression model was evaluated using accuracy, classification report, and confusion matrix.


```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy:", accuracy_score(y_test, y_pred_lr))

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_lr))

print("\nConfusion Matrix:\n")
print(confusion_matrix(y_test, y_pred_lr))
```

## Random Forest Model

Random Forest is an ensemble learning algorithm that builds multiple decision trees and combines their predictions.

Unlike Logistic Regression, which uses a linear decision boundary, Random Forest can capture complex non-linear relationships in network traffic data.

This model was implemented to improve detection performance and reduce false negatives in intrusion detection.


```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(n_estimators=100, random_state=42)

model_rf.fit(X_train, y_train)

y_pred_rf = model_rf.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_rf))

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_rf))

print("\nConfusion Matrix:\n")
print(confusion_matrix(y_test, y_pred_rf))
```

### Confusion Matrix Visualization

To better understand model performance, the confusion matrix was visualized using a heatmap.

The confusion matrix provides a detailed breakdown of:

- True Negatives (Normal correctly classified)
- False Positives (Normal incorrectly classified as Attack)
- False Negatives (Attack incorrectly classified as Normal)
- True Positives (Attack correctly classified)

Visualizing the confusion matrix helps analyze classification errors more clearly and assess the model’s reliability in detecting malicious network traffic.


```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Compute confusion matrix for Random Forest
cm_rf = confusion_matrix(y_test, y_pred_rf)

plt.figure(figsize=(6,5))
sns.heatmap(cm_rf,
            annot=True,
            fmt='d',
            cmap='Blues',
            xticklabels=['Normal (0)', 'Attack (1)'],
            yticklabels=['Normal (0)', 'Attack (1)'])

plt.title("Confusion Matrix - Random Forest")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.tight_layout()
plt.show()
```

The heatmap shows that the Random Forest model produces very few false positives and false negatives, indicating strong classification performance.

In intrusion detection systems, reducing false negatives is especially critical, as undetected attacks may compromise network security.


## Feature Importance Analysis

Random Forest provides feature importance scores that indicate how much each feature contributes to the final classification decision.

Analyzing feature importance helps identify which network traffic characteristics play the most significant role in detecting malicious activity.


```python
import numpy as np
import pandas as pd

# Get feature importance scores
importances = model_rf.feature_importances_

# Create dataframe for visualization
feature_importance_df = pd.DataFrame({
    'Feature': X_encoded.columns,
    'Importance': importances
})

# Sort features by importance
feature_importance_df = feature_importance_df.sort_values(
    by='Importance',
    ascending=False
)

# Display top 10 features
feature_importance_df.head(10)
```

### Feature Importance Analysis

Random Forest provides feature importance scores,
which indicate how much each feature contributes
to the classification decision.

Analyzing important features helps understand
which network traffic attributes are most relevant
for detecting malicious activity.


```python
plt.figure(figsize=(8,6))
sns.barplot(
    x='Importance',
    y='Feature',
    data=feature_importance_df.head(10)
)

plt.title("Top 10 Important Features - Random Forest")
plt.tight_layout()
plt.show()
```
